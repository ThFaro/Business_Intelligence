```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Dados
vendas_A = np.array([15, 22, 18, 20, 25, 17, 21, 24, 30, 27, 29, 23, 19, 21, 26, 22, 20, 24, 27, 30, 28, 25, 23, 19, 21, 22, 24, 29, 26, 31])
vendas_B = np.array([10, 12, 14, 15, 13, 11, 16, 18, 20, 17, 19, 15, 13, 12, 14, 16, 17, 18, 20, 22, 21, 19, 18, 16, 14, 12, 11, 15, 13, 16])

# Estatísticas descritivas
media_A = np.mean(vendas_A)
mediana_A = np.median(vendas_A)
moda_A = stats.mode(vendas_A).mode
variancia_A = np.var(vendas_A, ddof=1)
desvio_padrao_A = np.std(vendas_A, ddof=1)

media_B = np.mean(vendas_B)
mediana_B = np.median(vendas_B)
moda_B = stats.mode(vendas_B).mode
variancia_B = np.var(vendas_B, ddof=1)
desvio_padrao_B = np.std(vendas_B, ddof=1)

# Gráficos
plt.figure(figsize=(14, 7))

# Gráfico de dispersão
plt.subplot(1, 2, 1)
plt.scatter(range(len(vendas_A)), vendas_A, color='blue', label='Produto A', alpha=0.7)
plt.scatter(range(len(vendas_B)), vendas_B, color='green', label='Produto B', alpha=0.7)
plt.xlabel('Dia do Mês')
plt.ylabel('Unidades Vendidas')
plt.title('Gráfico de Dispersão das Vendas Diárias')
plt.legend()

# Gráfico de barras
plt.subplot(1, 2, 2)
barras = ['Média', 'Mediana', 'Moda']
altura_A = [media_A, mediana_A, moda_A]
altura_B = [media_B, mediana_B, moda_B]
x = np.arange(len(barras))  # Localização das barras

plt.bar(x - 0.2, altura_A, 0.4, label='Produto A')
plt.bar(x + 0.2, altura_B, 0.4, label='Produto B')
plt.xlabel('Estatísticas')
plt.ylabel('Valor')
plt.title('Estatísticas Descritivas das Vendas Diárias')
plt.xticks(x, barras)
plt.legend()

plt.tight_layout()
plt.show()

# Impressão dos resultados
print(f"Produto A - Média: {media_A:.2f}, Mediana: {mediana_A:.2f}, Moda: {moda_A}, Variância: {variancia_A:.2f}, Desvio Padrão: {desvio_padrao_A:.2f}")
print(f"Produto B - Média: {media_B:.2f}, Mediana: {mediana_B:.2f}, Moda: {moda_B}, Variância: {variancia_B:.2f}, Desvio Padrão: {desvio_padrao_B:.2f}")

```

#### 1. Gráfico de Caixa (Box Plot)
Um gráfico de caixa é útil para visualizar a distribuição dos dados, identificar outliers e comparar a dispersão entre diferentes grupos.

**Interpretação:** O gráfico de caixa mostra a mediana, os quartis e possíveis outliers das vendas diárias para cada produto. Pode ajudar a comparar a variabilidade e a presença de outliers entre os produtos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gráfico de caixa
plt.figure(figsize=(8, 6))
plt.boxplot([vendas_A, vendas_B], labels=['Produto A', 'Produto B'])
plt.title('Distribuição das Vendas Diárias por Produto')
plt.ylabel('Unidades Vendidas')
plt.show()
```

#### 2. Histograma

Um histograma é excelente para mostrar a distribuição de frequências de um conjunto de dados.

**Interpretação:** O histograma mostra a distribuição das vendas diárias em intervalos. Ele pode ajudar a identificar a forma da distribuição, como se é simétrica, enviesada ou multimodal.

```python
plt.figure(figsize=(12, 6))

# Histograma para Produto A
plt.subplot(1, 2, 1)
plt.hist(vendas_A, bins=10, color='blue', edgecolor='black')
plt.title('Histograma das Vendas do Produto A')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Frequência')

# Histograma para Produto B
plt.subplot(1, 2, 2)
plt.hist(vendas_B, bins=10, color='green', edgecolor='black')
plt.title('Histograma das Vendas do Produto B')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()

```

#### 3. Gráfico de Linha

Um gráfico de linha é útil para visualizar a tendência dos dados ao longo do tempo.

**Interpretação:** O gráfico de linha mostra como as vendas mudaram ao longo do mês para cada produto. É útil para identificar tendências, sazonalidades ou padrões.

```python
dias = np.arange(1, 31)  # Dias do mês

plt.figure(figsize=(10, 6))
plt.plot(dias, vendas_A, marker='o', linestyle='-', color='blue', label='Produto A')
plt.plot(dias, vendas_B, marker='o', linestyle='-', color='green', label='Produto B')
plt.xlabel('Dia do Mês')
plt.ylabel('Unidades Vendidas')
plt.title('Tendência das Vendas Diárias por Produto')
plt.legend()
plt.grid(True)
plt.show()
```

#### 4. Gráfico de Dispersão com Linha de Regressão

Adicionar uma linha de regressão a um gráfico de dispersão pode ajudar a visualizar a relação entre duas variáveis.

**Interpretação:** O gráfico de dispersão com linha de regressão ajuda a visualizar a tendência central das vendas ao longo do mês e comparar as tendências entre os dois produtos.

Esses gráficos adicionais proporcionam uma análise mais completa dos dados e permitem visualizar diferentes aspectos das vendas. Eles ajudam a responder perguntas específicas sobre a distribuição, tendência e comparação dos dados. Se precisar de mais informações ou de outro tipo de gráfico, é só avisar!

```python
from sklearn.linear_model import LinearRegression

# Dados para regressão
x = np.arange(len(vendas_A)).reshape(-1, 1)  # Dias do mês como variável independente
y_A = vendas_A
y_B = vendas_B

# Ajustar o modelo de regressão
model_A = LinearRegression().fit(x, y_A)
model_B = LinearRegression().fit(x, y_B)
linha_A = model_A.predict(x)
linha_B = model_B.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y_A, color='blue', label='Produto A')
plt.plot(x, linha_A, color='blue', linestyle='--', label='Regressão A')
plt.scatter(x, y_B, color='green', label='Produto B')
plt.plot(x, linha_B, color='green', linestyle='--', label='Regressão B')
plt.xlabel('Dia do Mês')
plt.ylabel('Unidades Vendidas')
plt.title('Vendas Diárias com Linha de Regressão')
plt.legend()
plt.grid(True)
plt.show()
```
