```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Dados
vendas_A = np.array([15, 22, 18, 20, 25, 17, 21, 24, 30, 27, 29, 23, 19, 21, 26, 22, 20, 24, 27, 30, 28, 25, 23, 19, 21, 22, 24, 29, 26, 31])
vendas_B = np.array([10, 12, 14, 15, 13, 11, 16, 18, 20, 17, 19, 15, 13, 12, 14, 16, 17, 18, 20, 22, 21, 19, 18, 16, 14, 12, 11, 15, 13, 16])

# Estatsticas descritivas
media_A = np.mean(vendas_A)
mediana_A = np.median(vendas_A)
moda_A = stats.mode(vendas_A).mode
variancia_A = np.var(vendas_A, ddof=1)
desvio_padrao_A = np.std(vendas_A, ddof=1)

media_B = np.mean(vendas_B)
mediana_B = np.median(vendas_B)
moda_B = stats.mode(vendas_B).mode
variancia_B = np.var(vendas_B, ddof=1)
desvio_padrao_B = np.std(vendas_B, ddof=1)

# Grficos
plt.figure(figsize=(14, 7))

# Grfico de disperso
plt.subplot(1, 2, 1)
plt.scatter(range(len(vendas_A)), vendas_A, color='blue', label='Produto A', alpha=0.7)
plt.scatter(range(len(vendas_B)), vendas_B, color='green', label='Produto B', alpha=0.7)
plt.xlabel('Dia do Ms')
plt.ylabel('Unidades Vendidas')
plt.title('Grfico de Disperso das Vendas Dirias')
plt.legend()

# Grfico de barras
plt.subplot(1, 2, 2)
barras = ['Mdia', 'Mediana', 'Moda']
altura_A = [media_A, mediana_A, moda_A]
altura_B = [media_B, mediana_B, moda_B]
x = np.arange(len(barras))  # Localizao das barras

plt.bar(x - 0.2, altura_A, 0.4, label='Produto A')
plt.bar(x + 0.2, altura_B, 0.4, label='Produto B')
plt.xlabel('Estatsticas')
plt.ylabel('Valor')
plt.title('Estatsticas Descritivas das Vendas Dirias')
plt.xticks(x, barras)
plt.legend()

plt.tight_layout()
plt.show()

# Impresso dos resultados
print(f"Produto A - Mdia: {media_A:.2f}, Mediana: {mediana_A:.2f}, Moda: {moda_A}, Varincia: {variancia_A:.2f}, Desvio Padro: {desvio_padrao_A:.2f}")
print(f"Produto B - Mdia: {media_B:.2f}, Mediana: {mediana_B:.2f}, Moda: {moda_B}, Varincia: {variancia_B:.2f}, Desvio Padro: {desvio_padrao_B:.2f}")

```

#### 1. Grfico de Caixa (Box Plot)
Um grfico de caixa  til para visualizar a distribuio dos dados, identificar outliers e comparar a disperso entre diferentes grupos.

**Interpretao:** O grfico de caixa mostra a mediana, os quartis e possveis outliers das vendas dirias para cada produto. Pode ajudar a comparar a variabilidade e a presena de outliers entre os produtos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Grfico de caixa
plt.figure(figsize=(8, 6))
plt.boxplot([vendas_A, vendas_B], labels=['Produto A', 'Produto B'])
plt.title('Distribuio das Vendas Dirias por Produto')
plt.ylabel('Unidades Vendidas')
plt.show()
```

#### 2. Histograma

Um histograma  excelente para mostrar a distribuio de frequncias de um conjunto de dados.

**Interpretao:** O histograma mostra a distribuio das vendas dirias em intervalos. Ele pode ajudar a identificar a forma da distribuio, como se  simtrica, enviesada ou multimodal.

```python
plt.figure(figsize=(12, 6))

# Histograma para Produto A
plt.subplot(1, 2, 1)
plt.hist(vendas_A, bins=10, color='blue', edgecolor='black')
plt.title('Histograma das Vendas do Produto A')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Frequncia')

# Histograma para Produto B
plt.subplot(1, 2, 2)
plt.hist(vendas_B, bins=10, color='green', edgecolor='black')
plt.title('Histograma das Vendas do Produto B')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Frequncia')

plt.tight_layout()
plt.show()

```

#### 3. Grfico de Linha

Um grfico de linha  til para visualizar a tendncia dos dados ao longo do tempo.

**Interpretao:** O grfico de linha mostra como as vendas mudaram ao longo do ms para cada produto.  til para identificar tendncias, sazonalidades ou padres.

```python
dias = np.arange(1, 31)  # Dias do ms

plt.figure(figsize=(10, 6))
plt.plot(dias, vendas_A, marker='o', linestyle='-', color='blue', label='Produto A')
plt.plot(dias, vendas_B, marker='o', linestyle='-', color='green', label='Produto B')
plt.xlabel('Dia do Ms')
plt.ylabel('Unidades Vendidas')
plt.title('Tendncia das Vendas Dirias por Produto')
plt.legend()
plt.grid(True)
plt.show()
```

#### 4. Grfico de Disperso com Linha de Regresso

Adicionar uma linha de regresso a um grfico de disperso pode ajudar a visualizar a relao entre duas variveis.

**Interpretao:** O grfico de disperso com linha de regresso ajuda a visualizar a tendncia central das vendas ao longo do ms e comparar as tendncias entre os dois produtos.

Esses grficos adicionais proporcionam uma anlise mais completa dos dados e permitem visualizar diferentes aspectos das vendas. Eles ajudam a responder perguntas especficas sobre a distribuio, tendncia e comparao dos dados. Se precisar de mais informaes ou de outro tipo de grfico,  s avisar!

```python
from sklearn.linear_model import LinearRegression

# Dados para regresso
x = np.arange(len(vendas_A)).reshape(-1, 1)  # Dias do ms como varivel independente
y_A = vendas_A
y_B = vendas_B

# Ajustar o modelo de regresso
model_A = LinearRegression().fit(x, y_A)
model_B = LinearRegression().fit(x, y_B)
linha_A = model_A.predict(x)
linha_B = model_B.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y_A, color='blue', label='Produto A')
plt.plot(x, linha_A, color='blue', linestyle='--', label='Regresso A')
plt.scatter(x, y_B, color='green', label='Produto B')
plt.plot(x, linha_B, color='green', linestyle='--', label='Regresso B')
plt.xlabel('Dia do Ms')
plt.ylabel('Unidades Vendidas')
plt.title('Vendas Dirias com Linha de Regresso')
plt.legend()
plt.grid(True)
plt.show()
```
